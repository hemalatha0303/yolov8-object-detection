```python
# Path to label file
label_file = "/content/runs/detect/train22/labels/car_and_dog (1).txt"

# Read label file and convert class IDs to names
with open(label_file, "r") as f:
    lines = f.readlines()

# Print results with object names
for line in lines:
    parts = line.strip().split()  # Split each line by space
    class_id = int(parts[0])  # Get class ID
    label_name = coco_classes.get(class_id, "Unknown")  # Convert to name
    bbox = parts[1:]  # Bounding box coordinates
    print(f"Detected: {label_name}, Bounding Box: {bbox}")
```
